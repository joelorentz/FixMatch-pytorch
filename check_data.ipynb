```python
import os
import csv
```

```python
u_root = "/home/joe/data/datasets/cebi/unlabeled"
csv_path = "/home/joe/data/datasets/cebi/unlabeled/init.csv"

date_dirs = [x for x in os.listdir(u_root) if x[0].isnumeric()] # get all dirs that are named as a date
print(date_dirs)

with open(csv_path,"w+") as f:
    writer = csv.writer(f, delimiter=",")
    writer.writerow(["sample_id","view0_path","view1_path","view2_path"])
    for dir in date_dirs:
        print(dir)
        date_path = os.path.join(u_root,dir)
        samples = os.listdir(date_path)
        print(len(samples))

        for sample in samples:
            viewpaths = [os.path.join(dir,sample,"%s.jpg"%x) for x in range(3)]
            row = [sample]
            row.extend(viewpaths)
            writer.writerow(row)
    

```
